```python
import matplotlib.pyplot  as plt

counts = emp_df['Attrition'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Attrition percentage')
```

```python
counts = emp_df['Department'].value_counts()
plt.figure(figsize=(7, 7))
plt.bar(counts.index, counts.values, color='steelblue')
plt.title('Employee Count per Department')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = emp_df.select_dtypes(include='number').corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
